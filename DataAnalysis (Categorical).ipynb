```python
df = pd.read_csv('C:/Users/ashim/Downloads/dataMobileCleaned.csv')
df.head()
```

```python
df.isna().sum()
```

```python
# Remove the Bahasa Melayu translations in the rows
df['Gender'] = df['Gender'].str.extract(r'(Male|Female)', expand=False)
```

```python
df.head()
```

```python
# Pie chart for Gender Distribution
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff'])
plt.title("Gender Distribution")
plt.ylabel('')  # Removes the y-label for a cleaner look
plt.tight_layout()
plt.show()


#The pie chart here shows the proportion of Male vs Female respondents in the survey. 
#It allows us to quickly visualize the gender balance of the user base. 
#A more equal split between male and female users may suggest a more generalized preference for mobile plans, while an imbalance could reveal insights into gender-specific trends in mobile usage.
```

```python
df['Occupation'] = df['Occupation'].str.extract(r'(Student|Employed|Unemployed|Retired)', expand=False)
df.head()
```

```python
# Verify the changes
df['Occupation'].value_counts()
```

```python
# Line plot for Occupation Distribution
occupation_counts = df['Occupation'].value_counts()
```

```python
# Bar chart for Occupation Distribution
plt.figure(figsize=(8, 4))
df['Occupation'].value_counts().plot(kind='bar', color='darkgreen')
plt.title("Occupation Distribution")
plt.xlabel("Occupation")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



#This line plot displays the distribution of respondents across different occupation categories. 
#It helps us identify which occupations are most prevalent in the survey population. 
#Understanding occupation trends can offer insights into how different job sectors may influence mobile plan preferences (e.g., students may prefer prepaid, while professionals may opt for postpaid).
```

```python
# Grouped bar chart for Gender vs Occupation
import matplotlib.pyplot as plt
occupation_gender = pd.crosstab(df['Occupation'], df['Gender'])
occupation_gender.plot(kind='bar', figsize=(10, 6), color=['#ff9999', '#66b3ff'])
plt.title("Gender vs Occupation")
plt.xlabel("Occupation")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['MaritalStatus'] = df['MaritalStatus'].str.extract(r'(Single|Married)', expand=False)
```

```python
# Stacked bar chart for Marital Status vs Type of Mobile Plan
import matplotlib.pyplot as plt
marital_plan = pd.crosstab(df['MaritalStatus'], df['MobilePlan'])
marital_plan.plot(kind='bar', stacked=True, color=['#f7c59f', '#77dd77'], figsize=(8, 4))
plt.title("Marital Status vs Type of Mobile Plan")
plt.xlabel("Marital Status")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
df['HaveWifi'] = df['HaveWifi'].str.extract(r'(Yes|No)', expand=False)
```

```python
df['Location'] = df['Location'].str.extract(r'(Rural|Sub-urban|Urban)', expand=False)
```

```python
#Pie charts for Wi-Fi availability in each location
import matplotlib.pyplot as plt
wifi_location = pd.crosstab(df['Location'], df['HaveWifi'])
wifi_location['Yes'].plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['yellow', 'orange', 'green'], figsize=(8, 6), labels=wifi_location.index)
plt.title("Wi-Fi Availability in Each Location")
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Remove bilingual values (anything after '/')
df = df.applymap(lambda x: x.split('/')[0].strip() if isinstance(x, str) and '/' in x else x)
```

```python
#Horizontal Bar charts for Contract Smartphone by Gender
import matplotlib.pyplot as plt
comparison = df.groupby(['Gender', 'HaveContractSmartphone']).size().unstack()
comparison.plot(kind='barh', figsize=(7, 4), color=['lightblue', 'lightgreen'])
plt.title("Contract Smartphone by Gender")
plt.xlabel("Respondents")
plt.ylabel("Gender")
plt.legend(title="Has Contract")
plt.tight_layout()
plt.show()
```

```python
# Grouped bar chart for hotspot usage by location
import matplotlib.pyplot as plt
hotspot_chart = df.groupby(['Location', 'HotspotUsage']).size().unstack()
hotspot_chart.plot(kind='bar', figsize=(8, 5), color=['#76c7c0', '#f7b267', '#f25c54'])
plt.title("Hotspot Usage by Location")
plt.xlabel("Location")
plt.ylabel("Number of Respondents")
plt.legend(title="Hotspot Usage")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
# Filter only respondents who answered "Yes" to PlayGames
gamers = df[df['Gaming'] == 'Yes']

# Count how many gamers are in each occupation
gamer_counts = gamers['Occupation'].value_counts()

# Plot pie chart
plt.figure(figsize=(7, 7))
colors = plt.cm.Paired(range(len(gamer_counts)))
plt.pie(gamer_counts, labels=gamer_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title("Distribution of Game Players by Occupation")
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
# Group and count by Location and EducationLevel
edu_location = df.groupby(['Location', 'EducationLevel']).size().unstack().fillna(0)

# Plot grouped bar chart
edu_location.plot(kind='bar', figsize=(10, 6), colormap='Set2')
plt.title("Education Level by Location")
plt.xlabel("Location")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=0)
plt.legend(title="Education Level", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
# Group by Occupation and SocialMedia
social_occ = df.groupby(['Occupation', 'SocialMedia']).size().unstack().fillna(0)

# Reorder columns if needed to ensure 'Yes' and 'No' are in correct order
social_occ = social_occ[['Yes', 'No']]

# Define custom colors: Green for Yes, Black for No
custom_colors = ['#2ecc71', '#000000']

# Plot
ax = social_occ.plot(kind='bar', figsize=(10, 6), color=custom_colors)

plt.title("Social Media Usage by Occupation")
plt.xlabel("Occupation")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45)
plt.legend(title="Social Media Usage", labels=['Yes', 'No'], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
