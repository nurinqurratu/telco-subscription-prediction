```python
import pandas as pd
dfCSV = pd.read_csv("C:/Users/nurin/Downloads/CSC649/Surveys.csv")
```

```python
dfCSV.info()
```

```python
#change column names
dfCSV.columns = ['Timestamp', 'MobilePlan', 'Satisfaction', 'Age', 'Gender', 'Occupation', 'Income', 'EducationLevel', 'MaritalStatus', 'Dependents', 'Location', 'MonthlyDataUsage', 'Gaming', 'Streaming', 'SocialMedia', 'AvgDailyScreenTime', 'HotspotUsage', 'MobileServiceProvider', 'YearsWithCurrentTelco', 'MonthlyBill', 'SIMCardsOwned', 'DeviceOwned', 'LinesInHousehold', 'HaveContractSmartphone', 'HaveWifi', 'Null']
dfCSV.info()
```

```python
#drop timestamp and null columns
dfCSV = dfCSV.drop(dfCSV.columns[[0, 25]], axis=1)
dfCSV.info()
```

```python
dfCSV.head()
```

```python
dfCSV.isna().sum()
```

```python
dfCSV.dropna(inplace = True)
```

```python
dfCSV.isna().sum()
```

```python
dfCSV.info()
```

```python
#convert columns to numeric
convertToNumeric = ['Income', 'Dependents']
for col in convertToNumeric:
    dfCSV[col] = pd.to_numeric(dfCSV[col], errors='coerce')
    dfCSV[col] = dfCSV[col].fillna(dfCSV[col].median())
dfCSV.info()
```

```python
#split columns with '/' and keep the first part
dfCSV = dfCSV.applymap(lambda x: x.split('/')[0].strip() if isinstance(x, str) and '/' in x else x)
dfCSV.head()
```

```python
#drop 'Satisfaction' column
dfCSV = dfCSV.drop(dfCSV.columns[[1]], axis=1)
dfCSV.info()
```

```python
#encode categorical variables
from sklearn.preprocessing import LabelEncoder
var_mod = ['MobilePlan', 'Gender', 'Occupation', 'EducationLevel', 'MaritalStatus', 'Location', 'Gaming', 'Streaming', 'SocialMedia', 'HotspotUsage', 'MobileServiceProvider', 'HaveContractSmartphone', 'HaveWifi']
le = LabelEncoder()
for i in var_mod:
    dfCSV[i] = le.fit_transform(dfCSV[i])
dfCSV.head()
```

```python
#Create a dataset file (after: drop missing data, transformation, drop columns )
dfCSV.to_csv(r'C:/Users/nurin/Downloads/CSC649/Project/project_cleaned', index = None, header=True)
```

```python
# Read data from external file 
import pandas as pd 
d = pd.read_csv("C:/Users/nurin/Downloads/CSC649/Project/project-Train.csv") 

d.info()
```

```python
#Data slicing 
x = d.iloc[:, 0:22] #features
y = d.iloc[:, 22:23] #label
```

```python
y.head()
```

```python
# Convert dataframe to array 
x = x.values 
y = y.values 
y = y.ravel() 
 
 
# K-Nearest Neighbor
# https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html
from sklearn.neighbors import KNeighborsClassifier 
ModelAKU = KNeighborsClassifier(n_neighbors=9, weights='distance', algorithm='auto', leaf_size=30, p=2, metric='minkowski') 

# Above ( ModelAKU = KNeighborsClassifier() ) same as below 
#ModelAKU = KNeighborsClassifier(n_neighbors=5, weights='uniform', algorithm='auto', leaf_size=30, p=2, metric='minkowski', metric_params=None, n_jobs=None) 


ModelAKU.fit(x, y) 
yPred = ModelAKU.predict(x) 
 
 
# Testing & performance analysis -Trainning data 
import sklearn.metrics as skm 
import matplotlib.pyplot as plt 
import seaborn as sns 
plt.figure(figsize=(9,9)) 
sns.heatmap(skm.confusion_matrix(y, yPred), annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r'); 
plt.ylabel('Actual label'); 
plt.xlabel('Predicted label'); 
all_sample_title = 'Trainning: Accuracy Score: {0}'.format(skm.accuracy_score(y, yPred)) 
plt.title(all_sample_title, size = 15);
```

```python
# Read data from external file 
import pandas as pd 
d2 = pd.read_csv("C:/Users/nurin/Downloads/CSC649/Project/project-Test.csv") 

d2.info()
```

```python
#Data slicing 
x2 = d2.iloc[:, 0:22] #features
y2 = d2.iloc[:, 22:23] #label

# Convert dataframe to array 
x2 = x2.values 
y2 = y2.values 
y2 = y2.ravel() 
 
yPred2 = ModelAKU.predict(x2) 
 
 
# Testing & performance analysis -Testing data 
import sklearn.metrics as skm 
import matplotlib.pyplot as plt 
import seaborn as sns 
plt.figure(figsize=(9,9)) 
sns.heatmap(skm.confusion_matrix(y2, yPred2), annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r'); 
plt.ylabel('Actual label'); 
plt.xlabel('Predicted label'); 
all_sample_title = 'Testing: Accuracy Score: {0}'.format(skm.accuracy_score(y2, yPred2)) 
plt.title(all_sample_title, size = 15);
```

```python
# Read data from external file 
import pandas as pd 
dfCSV = pd.read_csv("C:/Users/nurin/Downloads/CSC649/Project/project-Train.csv") 

dfCSV.info()
```

```python
dfCSV.head()
```

```python
import pandas as pd
from sklearn.model_selection import KFold

# Load your data into dataKFold
dataKFold = dfCSV

# Initialize KFold
kf = KFold(n_splits=4, shuffle=True, random_state=42)  # 4 splits for k-fold

# Create a counter for naming the files
fold_counter = 1

# Loop through the splits
for train_index, test_index in kf.split(dataKFold):
    # Split the data into training and testing sets
    train_data = dataKFold.iloc[train_index]
    test_data = dataKFold.iloc[test_index]
    
    # Save the training and testing data for this fold to separate files
    train_data.to_csv(f'C:/Users/nurin/Downloads/CSC649/Project/train_fold_{fold_counter}.csv', index=False)
    test_data.to_csv(f'C:/Users/nurin/Downloads/CSC649/Project/test_fold_{fold_counter}.csv', index=False)
    
    # Increment the fold counter
    fold_counter += 1
```
